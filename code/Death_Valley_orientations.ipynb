# Notebook for diabase sills in the Crystal Springs Formation in Death Valley bedding orientation compilation

Bedding orientations were taken in the field along the contact between the diabase sills and the Crystal Spring Formation sediments or on the sediments. Multiple measurements were taken on each sill. We average the measurements and use the mean orientation for tilt-correcting paleomagnetic dat. 

## importing external Python packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pmagpy.pmag as pmag
import pmagpy.ipmag as ipmag
import mplstereonet

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

## import data

```python
CS1_orientations = pd.read_csv('../data/orientations/CS1_orientations.csv')
CS2_orientations = pd.read_csv('../data/orientations/CS2_orientations.csv')
CS3_orientations = pd.read_csv('../data/orientations/CS3_orientations.csv')
CS4_orientations = pd.read_csv('../data/orientations/CS4_orientations.csv')
CS5_orientations = pd.read_csv('../data/orientations/CS5_orientations.csv')
CS6_orientations = pd.read_csv('../data/orientations/CS6_orientations.csv')
CS7_orientations = pd.read_csv('../data/orientations/CS7_orientations.csv')
CS8_9_orientations = pd.read_csv('../data/orientations/CS8_9_orientations.csv')
CS10_orientations = pd.read_csv('../data/orientations/CS10_orientations.csv')
CS11_orientations = pd.read_csv('../data/orientations/CS11_orientations.csv')
CS12_orientations = pd.read_csv('../data/orientations/CS12_orientations.csv')
CS13_orientations = pd.read_csv('../data/orientations/CS13_orientations.csv')
CS14_orientations = pd.read_csv('../data/orientations/CS14_orientations.csv')
CS15_orientations = pd.read_csv('../data/orientations/CS15_orientations.csv')
```

### Orientations for sill CS1

```python
CS1_orientations['pole_trend'] = CS1_orientations['dipAzimuth'] - 180
CS1_orientations['pole_plunge'] = 90 - CS1_orientations['dip']
CS1_mean_pole = ipmag.fisher_mean(CS1_orientations['pole_trend'].tolist(),CS1_orientations['pole_plunge'].tolist())
CS1_mean_strike = CS1_mean_pole['dec'] + 90
if CS1_mean_strike > 360:
    CS1_mean_strike = CS1_mean_strike - 360
CS1_mean_dip = 90 - CS1_mean_pole['inc']
print('mean strike: ', round(CS1_mean_strike, 1),  'mean dip: ', round(CS1_mean_dip, 1))
```

```python
fignum = 4
plt.figure(num=fignum,figsize=(5,5))
ipmag.plot_net(fignum)
ipmag.plot_di(CS1_orientations['pole_trend'].tolist(),CS1_orientations['pole_plunge'].tolist())
ipmag.plot_di_mean(CS1_mean_pole['dec'],CS1_mean_pole['inc'],CS1_mean_pole['alpha95'],color='red')
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111, projection='stereonet')
ax.plane(CS1_orientations['strike'], CS1_orientations['dip'],color='grey')
ax.plane(CS1_mean_strike, CS1_mean_dip,color='red',linewidth=5)
ax.pole(CS1_orientations['strike'], CS1_orientations['dip'],color='grey')
ax.pole(CS1_mean_strike, CS1_mean_dip,color='red')

plt.tight_layout()
plt.show()
print('sill CS1 mean tilt directions:', 'dip trend/dip:', round(CS1_mean_strike+90,1),  round(CS1_mean_dip, 1))
```

```python
def strike_to_dip_dir(strike_list):
    # assumption here is that what is getting fed into the function is a list of strike angles in degrees
    dip_direction_list = []
    
    for each_strike in strike_list:
        # to turn strike into dip direction: strike + 90
        each_dip_direction = each_strike + 90
        if each_dip_direction > 360:
            each_dip_direction = each_dip_direction - 360
        dip_direction_list.append(each_dip_direction)

    return dip_direction_list
```
